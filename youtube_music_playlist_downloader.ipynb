```python

# ===============================
# YouTube Music Playlist ‚Üí ZIP
# Single-cell Google Colab script
# ===============================

PLAYLIST_URL = "https://music.youtube.com/playlist?list=PLC5-GVt34RnhB8AFEZ-dmNKwggY_3y8ca"

# --- Install dependencies ---
!pip -q install yt-dlp
!apt-get -qq install ffmpeg zip

import os
from pathlib import Path
from google.colab import files

BASE_DIR = Path("/content/yt_music")
MUSIC_DIR = BASE_DIR / "music"
ZIP_PATH = BASE_DIR / "playlist.zip"

MUSIC_DIR.mkdir(parents=True, exist_ok=True)

print("‚ñ∂ Downloading playlist...")

!yt-dlp \
  --yes-playlist \
  -f bestaudio \
  --extract-audio \
  --audio-format mp3 \
  --audio-quality 0 \
  --embed-metadata \
  --embed-thumbnail \
  -o "/content/yt_music/music/%(playlist_index)02d - %(title)s.%(ext)s" \
  "$PLAYLIST_URL"

print("‚ñ∂ Creating ZIP...")

os.chdir(BASE_DIR)
!zip -r playlist.zip music

print("‚úî Done! Downloading ZIP...")
files.download(str(ZIP_PATH))
```
